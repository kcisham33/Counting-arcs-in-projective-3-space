Here are some functions we need.

```python
def generateAllSubsetsEvenSmall_p(copyLinearSpaces, numPoints):
    tbr = []
    for preSpace in copyLinearSpaces:
        gs = preSpace.ground_set()     #
        space = preSpace.blocks()
        n = numPoints
        while n > 4:
            for block in space:
                if len(block) == n:
                    for combo in Combinations(block, n-1):
                        if combo not in space:
                            space.append(combo)
                            #print combo
            n -= 1
        for n in xrange(4):                     #
            for combo in Combinations(gs, n):
                if combo not in space:
                    space += [combo]                #
        tbr.append(space)
    return tbr

def generateAllSubsetsEvenSmall_l(copyLinearSpaces, numPoints):
    tbr = []
    for preSpace in copyLinearSpaces:
        gs = preSpace.ground_set()     #
        space = preSpace.blocks()
        n = numPoints
        while n > 3:
            for block in space:
                if len(block) == n:
                    for combo in Combinations(block, n-1):
                        if combo not in space:
                            space.append(combo)
                            #print combo
            n -= 1
        for n in xrange(3):                     #
            for combo in Combinations(gs, n):
                if combo not in space:
                    space += [combo]                #
        tbr.append(space)
    return tbr

def bolster_l(f, numPoints):
    return IncidenceStructure(f.ground_set(),
                              generateAllSubsetsEvenSmall_l([IncidenceStructure(f.ground_set(),f.blocks())],numPoints)[0])

def bolster_p(f, numPoints):
    return IncidenceStructure(f.ground_set(),
                              generateAllSubsetsEvenSmall_p([IncidenceStructure(f.ground_set(),f.blocks())],numPoints)[0])


def autocard (A):
    automs = 0
    for a in A.isomorphic_substructures_iterator(A):
        automs += 1
    return automs

def autocard_both (A, B):
    automs = 0
    for a in A.isomorphic_substructures_iterator(A):
        if a in B.isomorphic_substructures_iterator(B):
            automs +=1
    return automs

def ways_to_extend_planes_bool (A, B, numPoints):
    A = copy.deepcopy(A)
    B = copy.deepcopy(B)
    A = bolster_p(A, numPoints)
    B = bolster_p(B, numPoints)
    inclusions = False
    for a in B.isomorphic_substructures_iterator(A):
        inclusions =True
        break

    return inclusions
   # return all_ways

def ways_to_extend_lines_bool (A, B, numPoints):
    A = copy.deepcopy(A)
    B = copy.deepcopy(B)
    A = bolster_l(A, numPoints)
    B = bolster_l(B, numPoints)
    inclusions = False
    for a in B.isomorphic_substructures_iterator(A):
        inclusions =True
    return inclusions

def ways_to_extend_planes_poset (A, B, numPoints):
    A = copy.deepcopy(A)
    B = copy.deepcopy(B)
    A = bolster_p(A, numPoints)
    B = bolster_p(B, numPoints)
    all_ways = []
    inclusions = 0
    for a in B.isomorphic_substructures_iterator(A):
        all_ways.append(a)
        inclusions += 1

    return inclusions / autocard(B)
   # return all_ways

def ways_to_extend_lines_poset (A, B, numPoints):
    A = copy.deepcopy(A)
    B = copy.deepcopy(B)
    A = bolster_l(A, numPoints)
    B = bolster_l(B, numPoints)
    all_ways = []
    inclusions = 0
    for a in B.isomorphic_substructures_iterator(A):
        inclusions += 1
        all_ways.append(a)
    return inclusions/autocard(B)

def ways_to_extend_planes (A, B, numPoints):
    A = copy.deepcopy(A)
    B = copy.deepcopy(B)
    A = bolster_p(A, numPoints)
    B = bolster_p(B, numPoints)
    all_ways = []
    inclusions = 0
    for a in B.isomorphic_substructures_iterator(A):
        all_ways.append(a)
        inclusions += 1

    #return inclusions / autocard(B)
    return all_ways

def ways_to_extend_lines (A, B, numPoints):
    A = copy.deepcopy(A)
    B = copy.deepcopy(B)
    A = bolster_l(A, numPoints)
    B = bolster_l(B, numPoints)
    all_ways = []
    inclusions = 0
    for a in B.isomorphic_substructures_iterator(A):
        inclusions += 1
        all_ways.append(a)
    return all_ways

def ways_to_extend(A1, B1, numPoints):
    A = copy.deepcopy(A1)
    B = copy.deepcopy(B1)
    all_ways_p = ways_to_extend_planes(A[0],B[0], numPoints)
    all_ways_l = ways_to_extend_lines(A[1],B[1], numPoints)
    inclusions = 0
    for a in all_ways_p:
        for b in all_ways_l:
            if a == b:
                #print a
                inclusions +=1
   # print inclusions, autocard(B[1]), autocard(B[0]), autocard_both(B[0], B[1])
    l1 = 0
    for b in B[0].blocks():
        if l1 < len(b):
            l1 =len(b)
    l2 = 0
    for b in B[1].blocks():
        if l2 <len(b):
            l2 = len(b)
    l = max(l1, l2)-1
  #  if l < numPoints-1 and l >0:
   #     B[0] = B[0].induced_substructure([0..l])
    #    B[1] =B[1].induced_substructure([0..l])
      #  print B[0].blocks(), B[1].blocks() ,autocard(B[0]), autocard(B[1]) , l
       # print B[0], B[1], "new i.s."
    #print inclusions, autocard(B[0]), autocard(B[1])
    return inclusions/autocard_both(B[0],B[1])



def make_w2e_db(start, numPoints):
    output = ["Undef!!" for x in range(numPoints+1)]
    for n in range(start, numPoints+1):
        print n
        row = []
        counter = 0
        for L in hyperSpaceList[n - 1]:
            counter+=1
            if counter % 10 ==0:
                print counter
            #print L[0].blocks(), L[1].blocks()
            rown = []
            for M in hyperSpaceList[n - 1]:
                rown += [ways_to_extend(L, M, n)]
            row += [rown]
        output[n] = row
    return output

#The following method determines the necessary planes for each LSF on numPoints many points
#the two main axioms that must be followed are:
#(a) if two (or more) lines intersect, then the union of the lines forms a  plane
#(b) if l = [l1,l2,l3,...,ln] is a line and x is a point not on that line, then [l1,l2,l3,...,ln,x] is a plane
import copy
def nec_planes(hsl_l, numPoints):
  
    l = hsl_l
   
    nec_planes = []
    final_planes = []
        #print "the LSF is ", l.blocks()
    lblocks = len(l.blocks())
    for s in Subsets(range(lblocks)): #collecting subsets of the lines
        if len(s) > 1:
            current_set = [(l.blocks()[i]) for i in s]
            inter = list(reduce(set.intersection, [set(l_) for l_ in current_set]))
            if len(inter) > 0: #if these lines all intersect, we need the union of the lines to be a plane
                nec_planes.append(list(reduce(set.union, [set(l_) for l_ in current_set])))
                    
        if len(s) ==1:
            for pt in range(numPoints):
                if pt not in l.blocks()[s[0]]: #if we have a single line l and a pt not on l, then [l, pt] is a plane
                    new_p = list(set(l.blocks()[s[0]]).union(set([pt])))
                    nec_planes.append(new_p)
        #print nec_planes
        
        ## now we try to get rid of redundant info. If P, Q are necessary planes but P is a proper subset of Q, we remove P
    for p in range(len(nec_planes)):
        flag = True
        for q in range(len(nec_planes)):
            if p !=q:
                if set(nec_planes[p]).issubset(set(nec_planes[q])):
                    flag = False
        if flag:
            final_planes.append(nec_planes[p])
    

            
    return final_planes #we return the full list of planes needed to satisfy axioms (a), (b) for each LSF        
```
